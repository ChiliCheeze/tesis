```python
from stdGrade import grd_system  # Import grd_system from grade_system.py
from stdStudyHabits2 import StudyHabitsSurvey
from faker import Faker
import random
import pandas as pd

# Setting a seed so that results are consistent across different runs
FAKER_SEED = 42
RANDOM_SEED = 42

# Library for generating synthetic names 
fake = Faker('en_PH')
Faker.seed(FAKER_SEED)
random.seed(RANDOM_SEED)

# List for randomly picking year level 
yrLvl = [1, 2, 3, 4]

# Dataframe lists
tbl1 = []  # Student details and grades
tbl2 = []  # Study habits survey

# Create a random student and generate their grades
for _ in range(239):  # 239 total respondents
    # Generate synthetic student information and grades
    student = grd_system(fake.name(), random.choice(yrLvl))

    # Append student info and grades to tbl1
    tbl1.append({
        "Student Name": student.name,
        "Year Level": student.year,
        "Student Number": student.stdNum,
        "Grades": student.grades,
        "GWA": student.final_grade,
        "Failed Subjects": student.sub_failed
    })

    # Generate study habits for this student
    stdStudyHabits = StudyHabitsSurvey(student.name, student.year, total_respondents=239, answered_respondents=200)
    study_habits_data = stdStudyHabits.generate_study_habits_data()

    # Add student name and number to the study habits data
    study_habits_data['Student Name'] = student.name
    study_habits_data['Student Number'] = student.stdNum

    # Append to tbl2
    tbl2.append(study_habits_data)

# Convert tbl1 to a DataFrame (student details and grades)
table_1 = pd.DataFrame(tbl1)

# Convert tbl2 to a DataFrame (study habits), combining all rows
table_2 = pd.concat(tbl2, ignore_index=True)

# Print the student's details and grades
print(table_1)

# Print the study habits survey data with student name and number
print(table_2)
```

```python
table_2.to_csv('tbl_12.csv', index=False)
print('Done')
```
